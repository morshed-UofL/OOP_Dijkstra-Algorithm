```python
# Q8: Implement Djikstraâ€™s algorithm for finding the shortest path 
#(modify your code, add any classes/methods/attributes needed).

import heapq

class Node:
    def __init__(self, index):
        self.index = index
        self.adjacent_edges = []

    def add_edge(self, edge):
        self.adjacent_edges.append(edge)

class Edge:
    def __init__(self, start_node, end_node, weight):
        self.start_node = start_node
        self.end_node = end_node
        self.weight = weight

    def print_edge(self):
        print(f"Edge from {self.start_node.index} to {self.end_node.index} with weight {self.weight}")

class Graph:
    def __init__(self):
        self.nodes = []

    def add_node(self, node):
        self.nodes.append(node)

    def add_edge(self, start_index, end_index, weight):
        start_node = self.nodes[start_index]
        end_node = self.nodes[end_index]
        edge = Edge(start_node, end_node, weight)
        start_node.add_edge(edge)

    def dijkstra(self, start_index):
        distances = {node: float('infinity') for node in self.nodes}
        distances[self.nodes[start_index]] = 0
        priority_queue = [(0, self.nodes[start_index])]

        while priority_queue:
            current_distance, current_node = heapq.heappop(priority_queue)

            for edge in current_node.adjacent_edges:
                neighbor = edge.end_node
                weight = edge.weight
                distance = current_distance + weight

                if distance < distances[neighbor]:
                    distances[neighbor] = distance
                    heapq.heappush(priority_queue, (distance, neighbor))

        return distances

# Example usage
graph = Graph()

# Add nodes
for i in range(5):
    graph.add_node(Node(i))

# Add edges
graph.add_edge(0, 1, 2)
graph.add_edge(1, 2, 3)
graph.add_edge(1, 3, 1)
graph.add_edge(2, 4, 5)
graph.add_edge(3, 4, 8)

# Find shortest paths from node 0
distances = graph.dijkstra(0)
for node, distance in distances.items():
    print(f"Distance from 0 to {node.index}: {distance}")
```
